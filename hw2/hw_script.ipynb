```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
```

```python
words = open('news.txt').read()
```

```python
test = words[:200]
```

```python
lemmas = []
from pymystem3 import Mystem
m = Mystem()
```

```python
def get_lemmas(text):
    ans = m.analyze(text)
    words = []
    for an in ans:
        try:
            if an['analysis'] != []:
                analysis = an['analysis'][0]
                if 'имя' not in analysis['gr'] and 'фам' not in analysis['gr'] and 'сокр' not in analysis['gr'] and an['text'].isupper() is False:
                    lemma = analysis['lex']
                    words.append(lemma)
        except KeyError:
            pass
        
    return(words) 
```

```python
lemmas = get_lemmas(words)
```

```python
len(lemmas)
```

```python
lemmas[:10]
```

```python
stopwords = nltk.corpus.stopwords.words('russian')
stopwords += ['это', 'свой', 'который', 'млрд', 'сей']
stopwords[:10]
```

```python
finder = BigramCollocationFinder.from_words(lemmas)
finder.apply_freq_filter(5)
finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in stopwords)
scored_bigrams = finder.score_ngrams(bigram_measures.likelihood_ratio)
sorted_bigrams = sorted(scored_bigrams, key = lambda x: x[1], reverse = True)
sorted_bigrams
```

```python
finder.nbest(bigram_measures.student_t, 10)
```

```python
finder.nbest(bigram_measures.chi_sq, 10)
```

```python
finder.nbest(bigram_measures.raw_freq, 10)
```

```python
finder.nbest(bigram_measures.pmi, 10)
```

```python
finder = BigramCollocationFinder.from_words(lemmas, window_size=6)
finder.apply_freq_filter(5)
finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in stopwords)
scored_bigrams = finder.score_ngrams(bigram_measures.likelihood_ratio)
sorted_bigrams = sorted(scored_bigrams, key = lambda x: x[1], reverse = True)
sorted_bigrams
```

```python
finder.nbest(bigram_measures.pmi, 10)
```

В целом кажется, что если сделать нормальную предобработку, разные метрики (LR, t-score, частота) дают не слишком отличающийся результат, даже просто частота срабатывает прилично. Самые странные результаты - при использовании PMI и хи-квадрата, но они же и самые специфичные. Ещё кажется, что ширина окна почти не влияет на результаты в верху списка.
